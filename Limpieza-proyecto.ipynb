# **Limpieza de Datos**

```python
import numpy as np
import pandas as pd
import pyreadstat as py
```

## ***- Primera encuesta de Polivictimización 2017***

#### **Lectura del archivo**

```python
# Leemos el archivo SPSS (.sav)
resultado = py.read_sav("Base de datos - 1.sav")

df_1 = resultado[0]     # DataFrame
meta_1 = resultado[1]   # Metadata
```

```python
df_1.head()
```

#### **Filtrado de los Datos**

```python
# Columnas que queremos conservar
col_conservar_1 = [
    'region', 'comuna', 'P1_1', 'P1_2', 'P1_3',
    'PA_5', 'PA_5_1', 'PA_5_2', 'PA_5_3', 'PA_5_4',
    'PA_6', 'PA_6_1', 'PA_6_2', 'PA_6_3', 'PA_6_4', 'PA_6_5',
    'PA_7', 'PA_7_1', 'PA_7_2', 'PA_7_3', 'PA_7_4', 'PA_7_5',
    
    'PB_1', 'PB_1_1', 'PB_1_2', 'PB_1_3', 'PB_1_4',
    'PB_2', 'PB_2_1', 'PB_2_2', 'PB_2_3', 'PB_2_4',
    
    'PC_1', 'PC_1_1', 'PC_1_2', 'PC_1_3',
    'PC_2', 'PC_2_1', 'PC_2_2', 'PC_2_3',
    'PC_5', 'PC_5_1', 'PC_5_2', 'PC_5_3',

    'PD_1', 'PD_1_1', 'PD_1_2',
    'PD_3', 'PD_3_1', 'PD_3_2', 'PD_3_3', 'PD_3_4', 'PD_3_5',
    'PD_4', 'PD_4_1', 'PD_4_2', 'PD_4_3', 'PD_4_4',
    'PD_5', 'PD_5_1', 'PD_5_2', 'PD_5_3', 'PD_5_4',
    'PD_6', 'PD_6_1', 'PD_6_2', 'PD_6_3', 'PD_6_4', 'PD_6_5',
    'PD_7', 'PD_7_1', 'PD_7_2', 'PD_7_3', 'PD_7_4', 'PD_7_5', 'PD_7_6',

    'PG_1', 'PG_2', 'PG_3', 'PG_4', 'PG_6', 'PG_8', 'PG_9', 'PG_10',
]
```

```python
# Filtramos por las columnas de interés
df_filtrado_1= df_1[col_conservar_1].copy()
df_filtrado_1.head()
```

```python
print(df_filtrado_1.columns)
```

#### **Limpieza de los Datos**

##### - Valores Nulos

```python
df_filtrado_1.info()
```

Al revisar la cantidad de valores nulos en cada columna, se observa que varias presentan datos faltantes. Sin embargo, dado que se trata de respuestas de una encuesta, no es necesario eliminar estas filas ni asignar valores de manera arbitraria, ya que estos reflejan omisiones o decisiones de los encuestados de no responder. Por lo tanto, estos valores no representan errores en los datos, sino información válida sobre la falta de respuesta.

##### - Renombramiento de columnas

```python
# Cambiamos el nombre de algunas columnas
df_filtrado_1.rename(columns={
    'region': 'Codigo_region',
    'comuna': 'Comuna_colegio',
    'P1_1': 'Curso_actual',
    'P1_2': 'sexo',
    'P1_3': 'edad'
}, inplace=True)
```

```python
df_filtrado_1.head()
```

##### - Conversión de tipos de datos

```python
# Identificamos todas las columnas de tipo float
cols_float = df_filtrado_1.select_dtypes(include='float').columns
```

```python
# Reemplazamos los valores nulos por 0 (como no existen registros con valor 0, esto no altera la integridad de los datos)
df_filtrado_1[cols_float] = df_filtrado_1[cols_float].fillna(0)
```

```python
# Convertimos los datos de tipo float a int 
df_filtrado_1[cols_float] = df_filtrado_1[cols_float].astype(int)
```

```python
df_filtrado_1.head() 
```

```python
# Buscamos en qué label se encuentra la información del código - comuna
for key, val in meta_1.value_labels.items():
    if 'IQUIQUE' in str(val.values()):
        print(key)
        break
```

```python
# Extraer el diccionario de comunas
dicc_comunas = meta_1.value_labels['labels195']

# Reemplazar los códigos por los nombres
df_filtrado_1['Comuna_colegio'] = df_filtrado_1['Comuna_colegio'].map(dicc_comunas)

print(df_filtrado_1[['Comuna_colegio']].head())
```

```python
df_filtrado_1.head()
```
